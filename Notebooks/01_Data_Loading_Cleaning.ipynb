# Step 8: Clean Text Values

Survey datasets often contain inconsistencies in text values due to manual user input. These inconsistencies can affect data analysis and lead to incorrect results.

Common issues include:

- Leading or trailing spaces
- Multiple spaces between words
- Inconsistent capitalization
- Hidden whitespace characters

In this step, we will clean all text (categorical) columns by:

- Removing leading and trailing spaces
- Replacing multiple spaces with a single space
- Preserving the original letter casing (for now)

This ensures that values such as:

- `"  Python"`
- `"Python  "`
- `"Python"`

are treated as the same value.

> **Note:** We will not change the letter case (uppercase/lowercase) in this step because some values, such as "B.Tech", "AI/ML", or "SQL", should retain their original formatting.

```python
text_columns = df.select_dtypes(include= "object").columns 
```

```python
print("Text columns : ")
print(text_columns.tolist())
```

```python
for column in text_columns: 
    df[column] = (
        df[column]
        .astype(str) 
        .str.strip()
        .str.replace(r"\s+", " ", regex= True)  
    )
```

```python
df.head()
```

```python
for column in text_columns:
    has_spaces = df[column].str.contains(r"^\s|\s$", regex=True, na=False).any()

    if has_spaces:
        print(f"{column} → Still contains extra spaces")
    else: 
        print("All good")
```

```python
for column in text_columns:
    df[column] = (
        df[column]
        .where(df[column].isna(), df[column].str.strip())
        .where(df[column].isna(), lambda s: s.str.replace(r"\s+", " ", regex=True))
    )
```

# Step 9: Check Missing Values

Missing values are one of the most common data quality issues in survey datasets. They may occur when respondents skip questions or choose not to provide certain information.

Before handling missing values, it is essential to identify:

- Which columns contain missing values
- How many values are missing
- The percentage of missing values in each column

This assessment helps us choose the most appropriate treatment for each column rather than applying a single method to the entire dataset.

> **Note:** In this step, we are only identifying missing values. We will decide how to handle them in the next step.

```python
missing_values = df.isnull().sum() 
missing_values
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_percentage.round(2)
```

```python
# Create a missing value summary table

missing_summary = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Missing Percentage (%)": (
        df.isnull().sum() / len(df) * 100
    ).round(2)
})

missing_summary = missing_summary.sort_values(
    by="Missing Percentage (%)",
    ascending=False
)

missing_summary
```

```python
missing_summary[
    missing_summary["Missing Values"] > 0
]
```

```python
# Visualize missing values

import matplotlib.pyplot as plt

missing_summary["Missing Percentage (%)"].plot(
    kind="bar",
    figsize=(12,5)
)

plt.title("Missing Values by Column")
plt.xlabel("Columns")
plt.ylabel("Missing Percentage (%)")
plt.xticks(rotation=90)
plt.show()
```
